```python
def extractdpop(dloc):
    data = pd.read_csv(open(dloc, 'rb'))
    data = data[["기준일ID",'행정동코드','총생활인구수']]
    data.행정동코드 = data.행정동코드.apply(lambda d: d//1000)
    dpop = data.groupby('행정동코드')['총생활인구수'].sum()
    dpop = dpop.to_frame()
    return dpop
```

```python
for m in range(201701,201713):
    month = str(m)
    location = "data/LOCAL_PEOPLE_" + month
    for d in range(1,32):
        try:
            if (d // 10) == 0:
                day = "0" + str(d)
            else:
                day = str(d)
            dloc = location + "/LOCAL_PEOPLE_" + month + day + '.csv'
            if d == 1:
                dpop = extractdpop(dloc)
            else:
                dpop = pd.concat([dpop,extractdpop(dloc)], axis = 1)
        except FileNotFoundError:
            pass
    if m==201701:
        mpop = dpop
    else:
        mpop = pd.concat([mpop, dpop], axis = 1)
```

```python
#변수 3개

mean_danger_all = mean_danger_all.to_frame()
mean_ses_all = mean_ses_all.to_frame()
mean_night_all = mean_night_all.to_frame()
```

```python
# num_fam = surveyrep.groupby("gu")["FAM1"].value_counts(normalize=True).to_frame()



oneperson = {}
for gu, gudata in surveyrep.groupby("gu"):
    oneperson[gu] = (gudata.FAM1 == 1).sum() / len(gudata.FAM1)


oneperson = pd.DataFrame.from_dict(oneperson, orient = 'index')

```

```python
oneperson.columns = ["1인가구비율"]
```

```python
oneperson #변수 1개
```

```python
highinc = {}
for gu, gudata in surveyrep.groupby("gu"):
    highinc[gu] = (gudata.INC >= 11).sum() / len(gudata.INC)


highinc = pd.DataFrame.from_dict(highinc, orient = 'index')
```

```python
highinc.columns = ['월500이상']
```

```python
onepersonmonth = {}
for gu, gudata in surveyrep.groupby("gu"):
    onepersonmonth[gu] = ((gudata.FAM1 == 1)&((gudata.SQ0_3 == 3)|(gudata.SQ0_3 == 4)|(gudata.SQ0_3 == 3))).sum() / len(gudata.FAM1)


onepersonmonth = pd.DataFrame.from_dict(onepersonmonth, orient = 'index')
```

```python
onepersonmonth.columns = ['1인월세']
```
